```python
# 수학 모듈
# 수학 계산 기능을 제공하는 모듈
# 제곱근, 삼각함수, 원주율 등을 포함

import math
print(math.sqrt(16)) #16의 제곱근 계산
```

```python
# 날짜, 시간 모듈
# 날짜와 시간을 다루는 모듈
# 현재 시간 확인, 날짜 계산 할때 사용함

import datetime

print(datetime.datetime.now())
```

```python
# 시간 제어 모듈
# 시간 흐름을 제어하는 모듈
# 프로그램을 잠시 멈출때 사용

import time
time.sleep(3) #프로그램을 3초동안 멈춘뒤 실행
print("3초 후 실행") # 3초가 지난 뒤 출력이 된다
```

```python
#시스템 모듈
# 파이썬 슬행 환경을 다루는 모듈
# 버전 확인, 프로그램 종료할때 사용

import sys

#현재 사용중인 파이썬 버전 출력 확인
print(sys.version)
```

```python
#운영체제 모듈
# 운영체제 = OS 와 관련된 기능을 제공
# 파일, 폴더 경로 처리에 사용함

import os

#현재 작업중인 폴더 경로 출력
print(os.getcwd())
```

```python
# 데이터 형식 모듈
# JSON 형식으로 데이터를 변환
# what is a JSON ? : 데이터를 약속된 모양으로 적는 방식 ("key"::"value")
# JSON은 웹에서 서버로 데이터를 전달할때, API 파일 저장에 사용
# API ? : 프로그램끼리 대화하는 방법(규칙)

import json
data = {
    "name" : "jin",
    "age" : 22
}

print(json.dumps(data))
```

```python
# 정규 표현식 모듈
# 문자열 패턴 검사 모듈
# 형식 확인(아이디, 이메일 등)에 사용

import re

#문자열이 "hi"로 시작하는지 확인
print(bool(re.match('hi', 'hi python')))
```

```python
# CSV 파일 모듈
# ,로 구분하는 txt기반 파일형식
# CSV ? : 엑셀 파일을 읽고 쓰는 모듈
# 데이터 저장/분석 기초에 사용

import csv

# CSV 파일에 한 줄 데이터를 저장한다
with open('test.csv', 'w', newline='') as f:
    csv.writer(f).writerow(['name', 'score'])
```

```python
# 자료구조 모듈
# 자료구조를 편하게 쓰도록 도와준다
# 카운트, 큐, 딕셔너리 확장 기능 제공

from collections import Counter

# 문자열에서의 각 글자의 개수를 센다
print(Counter('apple'))
```
