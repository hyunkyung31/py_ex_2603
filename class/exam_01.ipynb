```python
# class : 객체를 표현하기 위한 문법의 일종

class Person : # class = 설계도 (어떤 객체를 만들기 위한)
    def Hello(self) : #method / 클래스 안에 종속 / self라는게 붙는게 특징
        print("Hi")

#변수    객체
jin = Person() #jin은 person의 인스턴스 (객체), jin한테 person을 조종할 수 있는 리모컨을 준거랑 같다고 보면 됨
jin.Hello() #인스턴스 method
```
