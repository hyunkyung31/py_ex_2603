```python
# 문제 : 실행되는 출력문에는 `참` 그렇지 않으면 `거짓` 이라고 적어주세요.
# 조건 : `참1`, `참2`, `거짓1` 이런 순으로 적어주세요.

# 개념시작
if ( True ):
  print("참")
    
if ( False ):
  print("거짓")
# 개념끝

# 문제시작
a = 10

# `==` => 같다.
if ( a == 10 ):
  print("참1")

# `!=` => 같지 않다.
if ( a != 10 ):
  print("거짓1")

if ( a > 10 ):
  print("거짓2")

if ( a >= 10 ):
  print("참2")

b = 10

if ( a == b ):
  print("참3")

# c 에는 true or false 이 담긴다.
c = a != b

if ( c ):
  print("false")

if ( c == False ):
  print("true")

# `not` => 반전
if ( not c ):
  print("true")

if ( not(not c) ):
  print("false")

d = True

if ( c != d ):
  print("true")

if ( 10 > 5 and -6 > -10 and 10 != 10 ):
  print("false")
# and : 그리고
# or : 또는
#and를 기준으로 양쪽 값 true여야지 true
# True and True => True
# True and Flase => False =>양쪽 값이 모두 참이어야 한다!!
# True or True => True
# True or False => False
# False or Flase => Flase 값이 된다 =>한쪽 값만 참이여도 참이다!!

if ( 10 < 5 and -6 < -10 and 10 != 10 or 10 > 5 ):
  print("true")
# 문제끝
```

```python
# 문제 : 입력받은 정수가 소수인지 아닌지 알려주는 함수를 구현해주세요.

def is_prime_number(num):
  if num == 1 :
    return False
  
  i = 2
  while i < num :
    if num % i == 0 :
      return False
    i += 1

  return True #이게 실행되면 소수다

print("3은 소수입니다 : {}".format(is_prime_number(3)))
print("4는 소수입니다 : {}".format(is_prime_number(4)))
print("5는 소수입니다 : {}".format(is_prime_number(5)))
print("6는 소수입니다 : {}".format(is_prime_number(6)))
print("7는 소수입니다 : {}".format(is_prime_number(7)))
print("1000은 소수입니다 : {}".format(is_prime_number(1000)))
```

```python
'''
문제 - 딕셔너리에 각 달의 마지막 날들을 한번에 담기
조건 : 1월 ~ 3월까지
조건 : 2월의 마지막은 28일로 함
조건 : 첫 데이터의 key는 "1월", value는 31
'''

# month_end_days = {
#     "1월" : 31,
#     "2월" : 28,
#     "3월" : 31
# }

# print(month_end_days)

#각 달의 마지막 날들을 반복문을 통해 담기
month_end_days = {}
for i in range(1, 13) :
    month = str(i) + "월"
    end_day = 31 #기본값
    if i == 2 :
        end_day = 28
    elif i in [4, 6, 9, 11] :
        end_day = 30
    month_end_days[month] = end_day

print(month_end_days) #과정을 보고 싶으면 for문 안에 print를 넣어서 보면 된다 #내어쓰기는 shit + tap

```

```python
month_end_days = {}

month_end_days_list_ver = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
#데이터가 12개 (실제로는 24개 =>인덱스 값까지 포함) 리스트 + 인덱스 값까지 가져오는게 이너머레이트
for i, end_day in enumerate(month_end_days_list_ver) : # i, end_day가 변수
  month = str(i+1) + "월"
  month_end_days[month] = end_day

print(month_end_days)
```
