```python
# 인스턴스 속성, 기능

class Person() :
    def __init__(self) : # 걍 이부분은 셋팅임
        self.hello = "안녕하세요." #인스턴스 변수
    
    def hi(self) :
        print(self.hello)

jin = Person()  # 이때 __init__ method는 실행됨
jin.hi() 
```

```python
class Person() :
    def __init__(self, name, age, hobby) : 
        self.hello = "안녕하세요." 
        self.name = name
        self.age = age
        self.age = hobby
    
    def introduce(self) :
        print(f"{self.hello}, 저는 {self.age}살 취미가 {self.hobby}인 {self.name}입니다.") #클래스 내부에서 접근

jin = Person('진', 22, '운동')
jin.introduce()

print(f"나이 : {jin.age}") #클래스 외부에서 접근할때는 jin으로 접근함
print(f"이름 : {jin.name}")
print(f"취미 : {jin.hobby}")
```

```python
#class 외부에서는 접근 불가, class 내부에서만 사용 가능한 private attribute
# __는 보안, 클래스 안에서 사용하기 펼할때도 있다


class Person() :
    def __init__(self, name, age, hobby, weight) : 
        self.hello = "안녕하세요." 
        self.name = name
        self.age = age
        self.age = hobby
        self.__weight = weight
    
    def diet(self, loss_weight) :
        self.__weight -= loss_weight
        print(self.__weight)

jin = Person('진', 22, '운동', 88)
jin.diet(10)
```
