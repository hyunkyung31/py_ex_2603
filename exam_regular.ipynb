```python
# 사용 시 정규표현식 re모듈을 불러와야함
import re

text = "안녕하세요. 저는 15살 입니다, 저는 안경이 15개 있습니다."

# r = row string  \(특수문자) ->혼동하지 말라고 적어주는거
# \d = 숫자
# +는 한개 이상
pattern= r"\d+" #찾고 싶은 규칙을 적은거
# findall(찾을 규칙, 대상 문자열)
# 문자열에서 패턴과 맞는 내용을 전부 찾아서 리스트로 변환
result = re.findall(pattern, text)
print(result)
```

```python
# 특정 형식 찾기
data = "test123@gmail.com"
email_pattern = r"\w+@\w+\.\w+"
#\w : 문자 또는 숫자
# + 1개 이상
# @ = @
# w. : 점(특수 문자이기 떄문에 \를 붙인다)
# 문자/숫자 + @ 문자/숫자 + . + 문자/숫자
emails = re.findall(email_pattern, data)
print(emails)
```

```python
#문자열 치환

test = "가격은 12000원 입니다."

#sub(찾을패턴, 바꿀문자, 대상문자열)
#문자열에서 특정 패턴을 찾아서 다른 문자로 바꾸는 함수
new_text = re.sub(r'\d', '', test)
print(new_text)
```

```python
# 정규식 vs 일반 문자열 처리

test = "몸무게 : 70kg, 키 : 170cm"

test.split(',') #쪼개면 리스트로 2개만들어지고 숫자를 나눠서 출력가능
numbs = re.findall(r'\d+', test)
print(numbs)
```
