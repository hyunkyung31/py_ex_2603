```python
import matplotlib.pyplot as plt

# 기본 스타일 적용
plt.style.use('default')

# y값만 전달, x축은 자동으로 0, 1, 2, 3 이 설정된다
# plt.plot([0, 1, 2, 3], [2, 4, 6, 8,]) 과 아래 함수 같음 앞이 x축 뒤가 y축
plt.plot([2, 4, 6, 8])
plt.show()
```

```python
x = [2, 4, 6, 8]
y = [1, 3, 5, 7]

plt.plot(x, y)
plt.show()
```

```python
fig = plt.figure()

# (행개수, 열개수, 위치값) / 큰 도화지에 값을 나누고 싶을때 fig를 쓴다
fig.add_subplot(1, 2, 1)
fig.add_subplot(1, 2, 2)
plt.show()
```

```python
fig = plt.figure()

#(행개수, 열개수, 위치값)
a1 = fig.add_subplot(1, 2, 1)
a2 = fig.add_subplot(2, 2, 2)

x = range(0, 10)
y = range(5, 15)
a1.plot(x, y, 'rd')

x = range(0, 50)
y = range(0, 50)
a2.plot(x, y, 'y^')

plt.show()
```

```python
import matplotlib
import matplotlib.font_manager as fm

# 간격을 가능한 최적의 수치로 자동 조정
fig = plt.figure(constrained_layout=True)

# 원하는 폰트 파일 경로 문자열로 입력
font_location = "C:\\Windows\\Fonts\\malgunbd.ttf"

# 지정한 폰트 파일 경로에서 폰트 추출
fontname = fm.FontProperties(fname=font_location).get_name()

# matplotlib의 전역 폰트 설정 변경
matplotlib.rc('font', family=fontname)

fig = plt.figure()

#(행개수, 열개수, 위치값)
a1 = fig.add_subplot(1, 2, 1)
a2 = fig.add_subplot(2, 2, 2)

x = range(0, 10)
y = range(5, 15)
a1.plot(x, y, 'rd')

x = range(0, 50)
y = range(0, 50)
a2.plot(x, y, 'y^')

# x축, y축 이름 지정
a1.set_xlabel('시간')
a1.set_ylabel('돈')

a2.set_xlabel('시간')
a2.set_ylabel('돈')

plt.show()
```

```python
fig = plt.figure()

# 그래프 전체 디자인 스타일을 ggplot 스타일로 변경
# ggplot 특징 
# - 배경에 격자(grid)존재
# - 색감이 부드럽고 깔끔
# - 데이터 비교가 쉬움
plt.style.use("ggplot")

# a = fig.add_subplot(111) 1행 1열짜리 영역에서 1번째 위치에 그래프 생성
a = fig.add_subplot(111) 
x = [3, 4] # 막대의 위치 (좌표 기준)
y = [1, 2] # 막대의 높이 (값 기준)

# x축 라벨 이름 (tick 이름)
labels = ["english", "math"]

# 각 막대에 적용할 색상
colors = ["red", "blue"]
a.bar(x, y, tick_label=labels, color=colors)

plt.show()
```

```python
fig = plt.figure()

# 그래프 전체 디자인 스타일을 ggplot 스타일로 변경
# ggplot 특징 
# - 배경에 격자(grid)존재
# - 색감이 부드럽고 깔끔
# - 데이터 비교가 쉬움
plt.style.use("ggplot")

a = fig.add_subplot(111) 
x = [3, 4] # 막대의 위치 (좌표 기준)
y = [1, 2] # 막대의 높이 (값 기준)

# x축 라벨 이름 (tick 이름)
labels = ["english", "math"]

# 각 막대에 적용할 색상
colors = ["red", "blue"]
a.barh(x, y, tick_label=labels, color=colors) #h를 붙이면 가로로 나온다

plt.show()
```

```python
fig = plt.figure()

# 그래프 전체 디자인 스타일을 ggplot 스타일로 변경
# ggplot 특징 
# - 배경에 격자(grid)존재
# - 색감이 부드럽고 깔끔
# - 데이터 비교가 쉬움
plt.style.use("ggplot")

a = fig.add_subplot(111) 
x = [3, 4] # 막대의 위치 (좌표 기준)
y = [1, 2] # 막대의 높이 (값 기준)

# x축 라벨 이름 (tick 이름)
labels = ["english", "math"]

# 각 막대에 적용할 색상
colors = ["red", "blue"]
a.bar(x, y, tick_label=labels, color=colors) #h를 붙이면 가로로 나온다

#표준 수치 공식
for i, v in enumerate(y) :
    a.text(v+2, i + 1.05, str(v)) 

plt.show()
```
