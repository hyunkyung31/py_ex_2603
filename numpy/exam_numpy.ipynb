```python
import numpy as np

arr1 = np.array([1, 2, 3, 4])
arr2 = np.array([[1, 2, 3, 4],
                 [5, 6, 7, 8]])
arr3 = np.array([[[1, 2, 3, 4], [5, 6, 7, 8]],
                 [[0.1, 0.2, 0.3, 0.4], [5, 6, 7, 8]]])

# ndim : 배열의 차원 수
print(arr1.ndim)
print(arr2.ndim)
print(arr3.ndim)

# shape : 배열의 형태(배열 구조를 튜플로 보여주는거)
# (총 개수, 행개수, 열개수)
print(arr1.shape) # 데이터 개수 4개
print(arr2.shape) # 2줄이며 각 줄에 4칸
print(arr3.shape) # 2장(층) 각 장마다 2줄, 각 줄에 4칸

# size : 배열 요소의 총 개수
print(arr1.size) # 총 4개
print(arr2.size) # 2차원배열 2*4개 = 8개
print(arr3.size) # 2장에 2행에 총 4개 = 16개

# dtype : 배열 요소의 데이터 타입
# 기본적으로 int로 나오는데 소수점이 하나라도 섞여있으면 float으로 나온다
print(arr1.dtype)
print(arr2.dtype)
print(arr3.dtype)

# itemsize : 배열 원소 하나가 차지하는 메모리 크기
print(arr1.itemsize)
print(arr2.itemsize)
print(arr3.itemsize)

# data : 배열의 실제 데이터가 저장된 메모리 버퍼(buffer 주소) / 메모리주소
print(arr1.data)
print(arr2.data)
print(arr3.data)
```

```python
'''ndArray T : 전치 행렬 (행/열을 서로 바꾸는 것)
'''

arr = np.array([[1, 2, 3],
               [4, 5, 6]])
print("원래 배열")
print(arr)

print("전치 배열") # 행과 열이 바뀐다
print(arr.T)
```

```python
# nbArray.flatten() : 1차원으로 펼치기

arr = np.array([[1, 2, 3],
               [4, 5, 6]])
print("원래배열")
print(arr)

print("1차원으로 펼친 배열")
print(arr.flatten())

# flatten()은 복사본을 만든다 / 비파괴적
flat = arr.flatten()
flat[0] = 100
print("배열 출력")
print(arr)
```

```python
# nbArray.reshape() : 배열 모양 바꾸기

arr = np.array([1, 2, 3, 4, 5, 6])
print("원래 배열")
print(arr)

# reshape 할 때는 원본 size가 같아야 함
arr2 = arr.reshape(2, 3) #2행 3열로 반환
print(arr2)
```

```python
# ndArray.sum() : 배열 합계 구하기
# axis : 어느 축을 기준으로 계산할 것인가
# axis 옵션
# axis = 0 : 열 (세로축) 방향 합
# axis = 1 : 행 (가로축) 방향 합

arr = np.array([[1, 2, 3],
               [4, 5, 6]])

print("전체 합 :", arr.sum())


arr = np.array([[1, 2, 3],
               [4, 5, 6]])
print("열 방향 합(axis=0):", arr.sum(axis=0))
print("행 방향 합(axis=1):", arr.sum(axis=1))
```

```python
#nbArray.mean() : 평균 구하기

arr = np.array([[1, 2, 3],
               [4, 5, 6]])

print("전체 평균 :", arr.mean())


arr = np.array([[1, 2, 3],
               [4, 5, 6]])
print("열 방향 평균(axis=0):", arr.mean(axis=0))
print("행 방향 평균(axis=1):", arr.mean(axis=1))
```

```python
# ndArray.min() : 배열 최솟값 구하기
# ndArray.max() : 배열 최댓값 구하기

# axis : 어느 축을 기준으로 계산할 것인가
# axis 옵션
# axis = 0 : 열 (세로축) 방향 합
# axis = 1 : 행 (가로축) 방향 합

arr = np.array([[1, 2, 3],
               [4, 5, 6]])

print("전체 최솟값 :", arr.min())
print("전체 최댓값 :", arr.max())

print("열 방향 최솟값(axis=0_ :", arr.min(axis=0))
print("행 방향 최솟값(axis=0_ :", arr.max(axis=1))

print("열 방향 최댓값(axis=0_ :", arr.min(axis=0))
print("행 방향 최댓값(axis=0_ :", arr.max(axis=1))
```

1) np.zeros(shape)
   - 지정한 shape(모양)으로 배열을 만들고, 모든 원소를 0으로 채움
   - 주로 초기 값이 0인 배열이 필요할 때 사용(누적합, 버퍼, 마스크) 등
   - shape은 튜플로 넣는다 : (행, 열) 또는 (층, 행, 열)

```python
np.zeros((3, 4)) # 3행 4열의 2차원 배열 생성, 근데 데이터가 전부 0
```

2) np.ones(shape)
    - 지정한 shape(모양)으로 배열을 만들고, 모든 원소를 1로 채움
    - 기준 값을 1로 시작하거나, 가중치/마스크의 기본 값으로 자주 사용

```python
np.ones((2, 3, 4)) # 모든 원소가 1인 3차원 배열 생성, 2층에 3행 4열
```

3) np.full(shape, full_value)
    - 지정한 shape(모양)으로 배열을 만들고, 모든 원소를 특정 값인 (full_value)로 채운다
    - 임의의 상수로 초기화 할 때 사용

```python
np.full((4, 4), 3) #모든 값을 3으로 채운 4행 4열 배열이 만들어진다
```

4) *_like 계열 : np.zeros_like / np.ones_like / np.full_like
  - like 말 그대로 다른 배열을 보고 똑같이 따라 만듦
  - 즉, 기존 배열의 shape을 가져와서 새 배열 생성

  * 장점
    - shape을 매번 직접 적을 필요가 없음 => 실수가 줄어든다
    - 기본적으로 dtype도 원본을 따라감 => 원본과 같은 타입으로 안전하게 생성 가능
    

```python
arr2 = np.array([[1, 2, 3, 4],
                [5, 6, 7, 8]])

np.zeros_like(arr2)
# array([[0, 0, 0, 0],
#        [0, 0, 0, 0]])
np.ones_like(arr2)
# array([[1, 1, 1, 1],
#        [1, 1, 1, 1]])
np.full_like(arr2, 2)
# array([[2, 2, 2, 2],
#        [2, 2, 2, 2]])
```

```python
import numpy as np

''' 
2) seed(숫자)
   - 난수 생성의 초기 상태 고정
   - 같은 seed를 주면 항상 같은 난수 결과가 나옴
'''
np.random.seed(1) #난수 생성기준을 1로 고정하겠다
np.random.randint(0, 10, (2,3))

np.random.seed(2) #난수 생성기준을 2로 고정하겠다
np.random.randint(0, 10, (2, 2, 3)) #0 이상 10 미만의 정수 생성, shape(2, 2, 3)은 2층 2행 3열 구조 3차원 배열 생성
```

```python
import numpy as np

'''
인덱싱(indexing) : 특정 위치에 있는 하나의 값을 가져오는 방법
슬라이싱(slicing) : 일정 범위를 잘라서 여러 값을 가져오는 방법
'''

arr1 = np.array([1, 2, 3, 4]) # 1차원 배열, shape = (4,)
arr2 = np.array([[1, 2, 3, 4],
                [5, 6, 7, 8]]) # 2차원 배열, shape = (2, 4)

# ===== 1차원 배열 인덱싱 =====
print("=====파이썬 기본 리스트 예시=====")
li = [1, 2, 3, 4]
print(li[0])

# ===== 1차원 배열 인덱싱 =====
print("=====1차원 배열 인덱싱=====")
print(arr[1])

# ===== 1차원 배열 슬라이싱=====
# 기본형태 : arr[시작 : 끝 : 보폭]
# 시작 포함, 끝은 제외
print("=====1차원 배열 슬라이싱=====")
print(arr1[1:3:]) #1번 인덱스부터 3번 인덱스 이전까지 // arr1[1], arr[2] 뽑아오기


# ===== 2차원 배열 인덱싱=====
# 기본 형태 : arr2[행, 열]
print("=====2차원 배열 인덱싱=====")
print(arr2[0,1]) #첫번째 행(0번), 두번째 열(1번)
print(arr2[1,2]) #두번째 행(1번), 세번째 열(2번)


# ===== 2차원 배열 슬라이싱=====
# 기본 형태 : arr2[행 범위, 열 범위]

print("=====2차원 배열 슬라이싱=====")
print(arr2[1:2, 1:3]) #행 : 1번부터 2번 직전 -> 1번만 / 열 :1번부터 3번 직전 -> 1번, 2번


'''
- Fancy인덱싱 : 인덱스 번호를 리스트/배열로 여러개 전달
'''

print("=====1차원 배열 fancy 인덱싱=====")
print(arr1[[0, 1, 3]])
#0번, 1번, 3번 인덱스를 한번에 선택
#결과 [1, 2, 4]

print("=====2차원 배열 fancy 인덱싱=====")
print(arr2[[0, 1], [1,3]])
# (0, 1)위치와 (1, 3)위치를 각각 선택
# => arr2[0,1] = 2
# -> arr2[1, 3] = 8
# 결과 [2, 8]

print("=====fancy 인덱싱 + 슬라이싱=====")
print(arr2[:, [0, 3]]) #행 아무것도 안쓴건 전체 선택 , 열은 0번과 3번만 선택

# 논리값 인덱싱 : True/False 조건으로 필터링
print("논리값 인덱싱")
print(arr1[[True, True, False, True]])
print(arr1[arr1 > 2])
```
