```python
from openpyxl import*

## ===딕셔너리로 값 생성 후 엑셀에 넣기===

item1 = {
    '품목' : '바나나',
    '가격' : '3000',
    '수량' : 1,
    '날짜' : '2026년 3월 10일'
}

item2 = {
    '품목' : '사과',
    '가격' : '1200',
    '수량' : 12,
    '날짜' : '2026년 3월 11일'
}

item3 = {
    '품목' : '커피',
    '가격' : '3500',
    '수량' : 10,
    '날짜' : '2026년 3월 12일'
}

item4 = {
    '품목' : '신발',
    '가격' : '50000',
    '수량' : 4,
    '날짜' : '2026년 3월 13일'
}

wb = Workbook()
ws = wb.active
ws['A1'] = '품목' #A열 1행 : 품목
ws['B1'] = '가격' #B열 1행 : 가격
ws['C1'] = '수량' #C열 1행 : 수량
ws['D1'] = '날짜' #D열 1행 : 날짜

items = [item1, item2, item3, item4]

i = 2
for item in items :
    ws['A' + str(i)] = item['품목']
    ws['B' + str(i)] = item['가격']
    ws['C' + str(i)] = item['수량']
    ws['D' + str(i)] = item['날짜']
    i += 1

wb.save('test2.xlsx')
```

```python
from openpyxl import *
from openpyxl.styles import PatternFill

# =====셀 배경 색 바꾸기======

wb = Workbook()
ws = wb.active

# 헤더 넣기
ws['A1'] = '영어' 
ws['B1'] = '수학' 
ws['C1'] = '국어' 
ws['D1'] = '과학' 

color = PatternFill(fill_type='solid', fgColor= '2074e8')

#반복문을 통해 셀의 배경색 바꾸기
for cell in ws[1]: #1행 전체
    cell.fill = color

wb.save('bg_color.xlsx')

#엑셀 파일 읽기~~~~~
print(ws['A1'].value)
#기본 색상코드
# 빨강 : FF0000
# 노랑 : FFFF00
# 초록 : 00FF00
# 파랑 : 0000FF
# 회색 : D9D9D9
```

```python
from openpyxl import *
from openpyxl.chart import BarChart, Reference

# ====차트 만들깅=====

wb = Workbook()
ws = wb.active

# 예제 데이터 입력 
ws.append(['Name', 'Socre']) #헤더 행
ws.append(['홍길동', '95'])
ws.append(['홍길순', '50'])
ws.append(['홍길딩', '73'])
ws.append(['홍길당', '89'])

# Bar Chart 객체 생성 (막대그래프)
chart = BarChart()

#차트에 사용할 데이터 범위 지정
# min_col = 2 ->B열(score)
# min_row = 1 ->1행까지
# max_rpw = 5 -> 5행까지
data = Reference(ws, min_col = 2, min_row = 1, max_row = 5)
chart.add_data(data, titles_from_data = True)
# titles_from_data 는 첫 행을 Series name(계열이름)으로 사용 = 범례로 사용
# 지정한 데이터를 차트에 추가

# 차트 제목 설정
chart.title = "students score"

# 차트를 워크시트 E5위치에 추가
ws.add_chart(chart, "E5")


wb.save('chart_exam.xlsx')
```
