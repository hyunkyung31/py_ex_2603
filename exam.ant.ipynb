```python
#어노테이션 타입을 설명하는 표지 정도로 생각하면 된당 , 코드 이해하기 더 쉽고 명확하게 할 수 있따
#어떤 타입인지 미리 적어두는 표시
#에러는 안나고 가독성 올려주는

#변수
age : int = 30
print(age)

# 어노테이션은 강제규칙이 아니다
age : int = "안녕"
print(age)

#함수
#a int, b에도 int 들어간다, 결과값은 int다 / 미리 써두는거
def add(a:int, b:int) -> int:
    return a + b

add(10, 20)
add('안녕', '반가워')

# 리스트     /int 들어갈거다
numbs : list[int] = [1, 2, 3]

#딕셔너리   / 키값에는 str, 밸류값에는 int 들어갈거야~
person : dict[str, int] ={
    "age" : 30,
    "height" : 180
}
print(numbs)
print(person)
```
