#Pandas
- 파이썬에서 데이터 분석과 데이터 처리를 위한 대표적인 라이브러리
- 주로 표 형태의 데이터 (엑셀, CSV, 데이터베이스 등)을 다루는데 사용한다
- 데이터과학, 머신러닝, 금융 분석, 통계 처리 등에서 매우 많이 사용됨

```python
import pandas as pd

''' 
DataFrame : 여러개의 열 (Column)으로 구성된 '표' 형태의 데이터
 - "열 이름" : [값들]
 - 각 열에는 열 이름이 있음
 - 각 행에는 자동으로 행 번호(index)가 붙는다
'''

# 딕셔너리 형태로 DataFrame을 만듦
df = pd.DataFrame(
    {
        "Name": [
            "Braund, Mr. Owen Harris",
            "Allen, Mr. William Henry",
            "Bonnell, Miss Elizabeth",
        ],
        "Age": [22, 35, 58],
        "Sex": ["male", "male", "female"],
    }
)

# 전체 출력
print(df)

print("==열 하나씩 출력==")
# 결과는 DataFrame이 아니라 "Serise"라는 1차원 자료구조로 나온다
print(df["Name"])
print(df["Age"])
print(df["Sex"])

# Series : 1차원 데이터 (열 하나)
# DataFrame : 여러개의 Series가 모여서 만들어진 2차원 표 형태
print("===type출력===")
print(type(df['Age']))
print(type(df))
```

```python
# 이름이 Age인 Series 생성

ages = pd.Series([22, 35, 44], name="Age")
print(ages)

# DataFram에서 특정 열 꺼내기
print(df["Age"])

# 기본 통계 함수
# max : 최댓값
print("최대나이 :", df["Age"].max())
# min : 최솟값
print("최대나이 :", df["Age"].min())
# mean : 평균값
print("최대나이 :", df["Age"].mean())

# 요약 통계 정보 보기
# describe : 숫자형 열에 대해 "요약 통계 정보"를 한번에 보내줌
print(df.describe())

'''
<결과값 해석>
count   3.000000 : 데이터 개수
mean   38.333333 : 평균
std    18.230012 : 표준 편차 (데이터가 평균에서 얼마나 퍼져있는지)
min    22.000000 : 최솟값
25%    28.500000 : 1차분위수(Q1)
50%    35.000000 : 중앙값(median)
75%    46.500000 : 3사분위수(Q3)
max    58.000000 : 최댓값
'''

# DataFrame의 전체 요약 정보
# info : DataFrame이 "구조 정보"를 보여줌
print(df.info())
```

DataFrame에서 특정 행을 선택하는 방법 2가지
1) iloc : "위치(position)"로 선택(0번째, 1번째 처럼 순서/번호로 접근)
2) loc : "라벨(label)"로 선택(index에 붙어있는 이름으로 접근)

*iloc
- iloc은 '인덱스 값'이 뭐든 상관없이, '몇번째 줄'이냐로 가져옴
ex) df.iloc[0] => 첫번째 행(0번위치)
결과는 '한 행'이므로 Series(1차원) 형태로 나온다

```python
print(df.iloc[0]) # 0번 위치 (첫번째행)
print(df.iloc[1]) # 1번 위치 (두번째행)
print(df.iloc[2]) # 2번 위치 (세번째행)
```

* loc
loc은 "index 라벨"로 접근한다
현재는 iloc과 loc의 결과가 같음
index를 "이름"으로 바꾸면 ex) Name열을 index로 지정
loc은 이름을 기준으로 접근할 수 있다
df.index = ["사람1", "사람2", "사람3"]
df.loc["사람1"]로 접근

```python
df.loc[0]
```

head() : 위쪽(처음부분) 몇 개의 행만 보여주는 함수
tail() : 아래쪽(마지막 부분) 몇 개의 행만 보여주는 함수

```python
# 상위 5개만 보여준다
print(df.head())
# head()에 숫자를 안쓰면 기본 값은 5
# 즉, 위에서부터 5개의 행을 보여줌
# 데이터가 5개보다 적으면 있는만큼만 출력

# 상위 2개만 보여준다
print(df.head(2))
# head(2)는 위에서부터 2개의 행만 출력
# 데이터가 클 때 일부분만 빠르게 확인하는 경우 자주 사용

# 하위 5개만 보여준다
print(df.tail())
# head()에 숫자를 안쓰면 기본 값은 5
# 즉, 아래(끝)에서부터 5개의 행을 보여줌

# 하위 2개만 보여준다
print(df.tail(2))
# tail(2)는 마지막에서부터 2개의 행만 출력
# 최근 데이터나 마지막 데이터 확인할 때 자주 사용
```

#조건 필터링 (Blloean Indexing)


```python
# 나이가 30이상인 사람만 보기
print(df[df["Age"] >= 30])
# 성별이 여성(female)인 행만 보기
print(df["Sex"] == "female")
```

```python
#새로운 열 추가
df["Survived"] = [0, 1, 1] #추가하는거 넣고 데이터 넣기
print(df)

# 열 이름 바꾸기 rename
df = df.rename(columns = {"Sex" : "Gender"})
print(df)

# 열 삭제하기 (axis=0 : 열, axis=1: 행) drop
df = df.drop("Survived", axis=1 )
print(df)
```

```python
# 결측치(NaN) 처리
# 결측치란 값이 비어있는 상태
# pandas에서는 보통 NaN(Not of Number)으로 표시
# None을 넣어도 pandas에서는 자동으로 NaN으로 변환시킴

print(df)
print(df.isnull()) #빈값이 있는지 알려주는거 False라는건 빈값이 없다

df2 = df.copy()
#copy() : 원본 df를 복제(복사본 만들기)

df2.loc[1, "Age"] = None
print(df2.isnull())
# df2의 1번 행(두번째 사람)의 Age열 값을 None으로 변경
# loc[행 라벨, 열이름]형태로 특정 위치의 값을 바꿈

print(df2.dropna(axis=0))
#dropna() : 결측치가 있는 부분 제거를 한다 1번 행이 걍 사라진거

print(df2.fillna(0))
# fillna(i) : 결측치를 i값으로 채우기
```

```python
# 특정 열만 따로 채우기
# 결측치를 평균 값으로 채우기
print(df2["Age"].fillna(df["Age"].mean))
# Age열만 선택해서 fillna 적용
# NaN을 평균 값으로 대체
# 데이터 분석에서 자주 쓰는 방식

# 결측치를 이전 값으로 채우기
# print(df2.fillna(method="ffill")) #<버전 종료됨
print(df2.ffill())

# ffill : foward fill => 바로 위(이전)행의 값으로 NaN을 채움
# 시간순 데이터(날짜 데이터)에서 많이 사용

# 결측치를 다음 값으로 채우기
# print(df2.fillna(method="bfill")) #<버전 종료됨
print(df2.bfill())
# bfill : backward fill -> 바로 아래 (다음 행)의 값으로 NaN을 채움
# ffill과 반대 방향
```
